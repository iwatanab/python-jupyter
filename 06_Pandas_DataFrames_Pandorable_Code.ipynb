## #1 Avoid Chain Indexing ']['

df.loc['Wastenaw']['Total Polulation'] This will pull the rows with the index 'Washtenaw' and only the 'Total Pollution' Column If you see a ][ you should think carefully about what you are doing.

- loc is label-based. This means that if you write loc[2], you are looking for the values of your DataFrame that have an index labeled 2.
- iloc is position-based. This means that if you write iloc[2], you are looking for the values of your DataFrame that are at index 2.

```python
#Remember, a dataframe is a a stack of series.
#The code below pulls the first row based on it's index
census.iloc[0].head()
```

```python
#The code below pulls the first row based on it's label
census.set_index(['STNAME','CTYNAME'], inplace = True)
census.loc[('Alabama','Autauga County')]
```

```python
#The following code pulls the DIVISON and CENSUS2010POP for Autauga County in Alabama
#Note the structure [Row Label, Column Label]
# .loc[ROW LABEL, COLUMN INDEX] will not work (e.g. census.loc[('Alabama','Autauga County'),[1:4] )
census.loc[('Alabama','Autauga County'),['DIVISION','CENSUS2010POP']]
```

## #3 Apply Functions to Rows in a DataFrame Using apply()

```python
# Say we want to get the minimum and maximum values for the population estimate across the years

import numpy as np

# Quick note: How to pull multiple columns from a DataFrame
# THIS WILL FAIL: census['POPESTIMATE2010','POPESTIMATE2011']
# You must wrap the multiple column names in a list
census[['POPESTIMATE2010','POPESTIMATE2011']].head()
```

```python
def min_max(row):
    data = row[['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']]
    return pd.Series({'min':np.min(data), 'max':np.max(data)})

aa = census.apply(min_max, axis=1).head() #Takes the series 
#returned for each row, flips them 90 degrees clockwise and stacks them
aa.head()
```

```python
# As you would expect, we can use lamba function in the apply method
rows = ['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']

census.apply(lambda x: np.max(x[rows]), axis=1).head()
```
